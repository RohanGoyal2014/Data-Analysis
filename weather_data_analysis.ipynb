```python
df2 = pd.read_csv('testset.csv')
```

```python
df2
```

```python
df2 = df2.loc[:,'datetime_utc':' _conds']
```

```python
df2
```

```python
df2['datetime_utc'] = pd.to_datetime(df2['datetime_utc'])
df2 = df2.set_index('datetime_utc')
```

```python
df2.head()
```

```python
weather_conditions = dict()
```

```python
for index,row in df2.iterrows():
    yr = index.year
    month = index.month
    cond = row[' _conds']
#     print(month,yr,cond)
    if yr == 1997 and month == 1:
        if cond not in weather_conditions:
            weather_conditions[cond]=0
        weather_conditions[cond]+=1
```

```python
plt.pie(list(weather_conditions.values()),labels=weather_conditions.keys())
```
